```python
import json
import numpy as np
from numpy.polynomial.polynomial import Polynomial
```

```python
model = json.load(open('model.json'))
```

```python
EPSI1 = model['epsi1']
EPSI21 = model['epsi21']
EPSI22 = model['epsi22']
EPSI3 = model['epsi3']
EPSI4 = model['epsi4']
```

```python
L1 = Polynomial(model['l1'])
R1 = Polynomial(model['r1'])

L2 = Polynomial(model['l2'])
C2 = Polynomial(model['c2'])
R2 = Polynomial(model['r2'])

L3 = Polynomial(model['l3'])
R3 = Polynomial(model['r3'])

L4 = Polynomial(model['l4'])
R4 = Polynomial(model['r4'])
```

```python
def a1(epsi):
    if epsi < EPSI1:
        return L1(epsi)
    else:
        return R1(epsi)
```

```python
def a2(epsi):
    if epsi < EPSI21:
        return L2(epsi)
    elif epsi < EPSI22:
        return C2(epsi)
    else:
        return R2(epsi)
```

```python
def a3(epsi):
    if epsi < EPSI3:
        return L3(epsi)
    else:
        return R3(epsi)
```

```python
def a4(epsi):
    if epsi < EPSI4:
        return L4(epsi)
    else:
        return R4(epsi)
```

```python
from matplotlib import pyplot as plt
```

```python
epsi = np.linspace(0.0, 0.45)
for i, ai in enumerate([a1, a2, a3, a4]):
    plt.plot(epsi, [ai(e) for e in epsi], '--')
    plt.xlabel(r'$\varepsilon$')
    plt.ylabel(fr'$a_{i + 1}$')
    plt.show()
```

```python
epsi = np.linspace(0.17, 0.40)
plt.plot(epsi, [a2(e) for e in epsi], '--')
plt.xlabel(r'$\varepsilon$')
plt.ylabel(r'$a_2$')
plt.show()
```
